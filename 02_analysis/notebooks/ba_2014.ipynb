```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon, shape
import geopandas as gpd
import pickle

# Import custom functions from `scripts` folder
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from scripts.clean_tweets import geometrize_tweets, convert_shapefile_crs, find_frequencies
from scripts.home_location import assign_home_location

np.random.seed(42)
pd.set_option('display.max_columns', 999)
```

# Loading data

```python
ba14 = pd.read_csv('../data/tweets/ba_2014.csv')
```

```python
ba_shapefiles = gpd.read_file('../data/shapefiles/buenos_aires_shapefiles/BA_2010.shp')
ba_shapefiles = convert_shapefile_crs(ba_shapefiles)
```

# Filtering data and adding tracts

```python
# Compute summary stats
num_tweets_14, num_users_14, median_tweets_14 = len(ba14), ba14['u_id'].nunique(), ba14.groupby('u_id').size().median()
```

```python
# Filter based on median tweets for 2014
ba14_filtered = ba14.groupby('u_id').filter(lambda group: len(group) >= median_tweets_14)
```

```python
# Geometrize tweets based on lat/lon
ba14_filtered = geometrize_tweets(ba14_filtered)
```

```python
# Add datetime
ba14_filtered['timestamp'] = pd.to_datetime(ba14_filtered['created_at'] // 1000, unit='s')
ba14_filtered['date'] = ba14_filtered['timestamp'].dt.date
ba14_filtered['hour'] = ba14_filtered['timestamp'].dt.hour
```

```python
# Spatial join for tracts; add home locations
ba14_filtered = gpd.sjoin(ba14_filtered, ba_shapefiles, how='left', op='intersects')
ba14_filtered['home'] = assign_home_location(ba14_filtered, tract='codigo')
```

```python
# Save data for easy loading in future
# with open('../data/ba14_home.pkl.', 'wb') as file:
#     pickle.dump(ba14_filtered, file, protocol=4)
with open('../data/ba14_home.pkl', 'rb') as file:
    ba14_filtered = pickle.load(file)
    
ba14_filtered['is_home'] = ba14_filtered['home'] == ba14_filtered['codigo']
```

```python
# How many of the filtered tweets have a home assignment?
print((~ba14_filtered['home'].isnull()).mean())
```

```python
# Ratio of users that have home location
ba14_filtered.groupby(['u_id']).first()['is_home'].mean()
```

```python
# Ratio of filtered 2014 tweets in which tweet was made from home tract 
ba14_filtered['is_home'].mean()
```

```python
### Plot 0.5% of 2014 tweets
# Blue tweets: Resident
# Red tweets: Non-resident
# Gray (barely visible beneath the blue): Buenos Aires shapefile
fig, ax = plt.subplots(figsize=(10, 7))
ba_shapefiles['geometry'].plot(ax=ax, color='gray')
smpl = ba14_filtered.sample(frac=0.005, random_state=42)
smpl[~smpl['is_home']].plot(ax=ax, marker='o', color='red', alpha=0.01, label='Red: Non-Resident')
smpl[smpl['is_home']].plot(ax=ax, marker='o', color='blue', alpha=0.01, label='Blue: Resident')
plt.legend()
plt.axis([-58.6, -58.3, -34.75, -34.5])
plt.title('Buenos Aires tweets, 2014');
```
